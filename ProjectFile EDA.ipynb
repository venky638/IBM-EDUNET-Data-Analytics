```python
# Load the data
df = pd.read_csv('SampleSuperstore.csv')

# Identify duplicate records
duplicate_rows = df[df.duplicated()]

# Remove duplicate records
df = df.drop_duplicates()
df
```

# Some Statistical Information 

```python
df.describe()
```

```python
# Finding the top 10 states by sales

top_5_States = df.groupby("State")["Sales"].sum().sort_values(ascending = False).head(10)

#plot the sales by States

plt.bar(top_5_States.index,top_5_States.values)
plt.xlabel("State")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```

```python
# Finding the top  states by Profit

top_5_States = df.groupby("State")["Profit"].sum().sort_values(ascending = False).head(10)

#plot the Top 10 profit States

plt.bar(top_5_States.index,top_5_States.values)
plt.xlabel("State")
plt.ylabel("Profit")
plt.xticks(rotation=45)
plt.show()
```

```python
# Finding the sales based on category

category_sales = df.groupby ("Category")["Sales"].sum()

#Creating a pie chart of distribution of sales by catergory
plt.pie(category_sales, labels=category_sales.index, colors=["#008000", "#FF0000", "#0000FF"],shadow = True,autopct="%2.1f%%")
plt.title("Distribution of Sales by Category")
plt.show()
```

```python
# Finding the sales based on category

category_Profit = df.groupby ("Category")["Profit"].sum()

#Creating a pie chart of distribution of sales by catergory
plt.pie(category_Profit, labels=category_Profit.index, colors=["#008000", "#FF0000", "#0000FF"],shadow = True,autopct="%2.1f%%")
plt.title("Profits made by each Category")
plt.show()
```

```python
# Group data by sub-category and calculate total sales
grouped_df = df.groupby('Sub-Category')['Sales'].sum()

# Sort the DataFrame by total sales in descending order
top_Subcategories = grouped_df.sort_values(ascending=False).head(10)

# Create a bar chart for top-selling sub-categories
plt.bar(top_Subcategories.index, top_Subcategories.values)
plt.xlabel('Sub-Category')
plt.ylabel('Total Sales')
plt.title('Top 10 Selling Sub-Categories')
plt.xticks(rotation=45)
plt.show()
```

```python
# Group data by sub-category and calculate total sales
grouped_df = df.groupby('Sub-Category')['Profit'].sum()

# Sort the DataFrame by total sales in descending order
top_Subcategories = grouped_df.sort_values(ascending=False).head(10)

# Create a bar chart for top-selling sub-categories
plt.bar(top_Subcategories.index, top_Subcategories.values)
plt.xlabel('Sub-Category')
plt.ylabel('Total Sales')
plt.title('Top 10 Profit Sub-Categories')
plt.xticks(rotation=45)
plt.show()
```

```python
# Group data by region and calculate total sales
grouped_df = df.groupby('Region')['Sales'].sum()

# Create a pie chart
plt.pie(grouped_df, labels=grouped_df.index, autopct='%1.1f%%',shadow = True,explode=[0,0,0,0.2])
plt.title('Sales Distribution by Region')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.show()
```

```python
# Group data by region and calculate total Profits
grouped_df = df.groupby('Region')['Profit'].sum()

# Create a pie chart
plt.pie(grouped_df, labels=grouped_df.index, autopct='%1.1f%%',explode=[0,0,0,0.2],shadow = True)
plt.title('profit made  by Regions')
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular.
plt.show()
```

```python
# Group data by region and category and calculate total sales
sales_by_region_category = df.groupby(['Region', 'Category'])['Sales'].sum()

# Reshape the data to have region-wise categories as columns
sales_by_region_category = sales_by_region_category.unstack()

# Create a stacked bar chart to visualize sales distribution by category for each region
plt.figure(figsize=(12, 6))
sales_by_region_category.plot(kind='bar', stacked=True)
plt.xlabel('Region')
plt.ylabel('Total Sales')
plt.title('Sales Distribution by Category for Each Region')
plt.legend(title='Category')
plt.xticks(rotation=0)
plt.show()
```
